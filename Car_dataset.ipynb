```python
import pandas as pd 
import numpy as np
```

```python
pd.set_option("display.max_columns",None)
```

```python
df = pd.read_csv("C:\\datasets\\Automobile_data.csv")
df.head()
```

```python
df.info()
```

```python
object_columns = df.select_dtypes(include=["object"]).columns
non_object_columns = df.select_dtypes(exclude=["object"]).columns
```

```python
len(object_columns),len(non_object_columns)
```

```python
df.replace("?",np.nan,inplace=True)
```

```python
df.head()
```

- Converting Object value to Numeric values

```python
null_obj_columns2 = [ 'bore','normalized-losses', 'stroke',
 'horsepower',
 'peak-rpm',
 'price']
```

```python
df[null_obj_columns2] = df[null_obj_columns2].apply(pd.to_numeric,errors='coerce')
```

```python
df.isna().sum()
```

```python
print("Object columns empty name: ")
null_obj_columns = []
empty=0
for col in object_columns:
    empty = df[col].isnull().sum()
    if empty>0:
        null_obj_columns.append(col)
        print(f"\t* {col} -> {empty}")
```

- Counting unique value in the categorical columns

```python
df["num-of-doors"].value_counts()
```

```python
df["num-of-doors"].fillna("four",inplace=True)
```

```python
print("Numerical columns empty name: ")
null_numerical_columns = []
empty=0
for col in non_object_columns:
    empty = df[col].isnull().sum()
    if empty>0:
        null_numerical_columns.append(col)
        print(f"\t* {col} -> {empty}")
```

- As we can observe that there are 6 categorical columns with null values
- 1 Categorical columns have empty values 

```python
null_numerical_columns.remove('price')
null_numerical_columns
```

```python
for col in null_numerical_columns:
    mean = round(df[col].mean(),2)
    df[col] = df[col].fillna(mean)
```

```python
# Separate rows with any NaN values 
df_with_nan = df[df.isna().any(axis=1)] 
# Separate rows without any NaN values 
df_without_nan = df[df.notna().any(axis=1)]
```
