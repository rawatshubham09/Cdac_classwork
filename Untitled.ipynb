```python
x = np.arange(0,4*np.pi,0.1)
y = np.sin(x)
z = np.cos(x)

fig, ax = plt.subplots()

ax.plot(x,y)
ax.plot(x,z)

ax.set_title("Two Trig Function")
ax.legend(["Sin","Cos"])
ax.xaxis.set_label_text("Angle 0")
ax.yaxis.set_label_text("Sine and Coseine")

ax.set_ylim(-1,1)

plt.show()
```

```python
df = pd.DataFrame(data=[100,20,500], index=["BMW","Lord Alto", "Splender"], columns=["Unit-Solds"])
```

```python
df
```

```python
plt.bar(df.index, df["Unit-Solds"],color=['r','b','g'])
plt.xlabel("Vehicles")
plt.ylabel("No. of unit Solds")
plt.title("Dummy-Set")
```

```python
## Another way to plot the graph

fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.bar(df.index, df["Unit-Solds"],color=['r','b','g'])
ax.xaxis.set_label("Vehicle")
ax.yaxis.set_label("no of units")
ax.title.set_label("Dummy")
```

```python
# Stack bar chart
x = np.arange(1)
BMW = 100
Splender = 500

plt.bar(x,BMW,width=0.35)
plt.bar(x,Splender,width=0.35,bottom=BMW)
plt.xlabel("Vehicle")
```

```python
import matplotlib.pyplot as plt
y = [3.2, 3.9, 3.7, 3.5, 3.02199]
x = [0.15, 0.3, 0.45, 0.6, 0.75]
n = [155, "outliner", 293, 230, 670]

fig, ax = plt.subplots()
ax.scatter(x, y)

for i, txt in enumerate(n):
    ax.annotate(txt, (x[i], y[i]))
plt.show()
```
