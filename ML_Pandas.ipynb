## Creating Random dataFrame

```python
import numpy as np
```

```python
temp = np.random.randint(low=20,high=100,size=[20])
name =np.random.choice(["Ram", "Shyam", "Rohit", "Rishib"],20)
random = np.random.choice(["Reliance","BSNL","IDEA","Uninore","JIO"],20)
```

```python
a = list(zip(temp,name,random))
a[:5]
```

```python
df = pd.DataFrame(data=a,columns=["Temp","Name","Random"])
df
```
