```python
X = df.drop(["target","flower_name"],axis=1)
y = df.target
```

```python
from sklearn.metrics import f1_score
```

```python
k_value = []
train_f1_score = []
test_f1_score = []

for i in range(1,20,2):
    model = KNeighborsClassifier(n_neighbors=i)
    model.fit(X_train,y_train)
    train_f1_score.append(f1_score(y_train,model.predict(X_train),average="micro"))
    test_f1_score.append(f1_score(y_test,model.predict(X_test),average="micro"))
    k_value.append(i)
```

```python
plt.plot(k_value,train_f1_score,color="green")
plt.plot(k_value,test_f1_score,color="red")
plt.legend(["Train Score","Test Score"])
plt.xlabel("Number of Neighbors")
plt.ylabel("F1_Score")
plt.title("Test vs Train Accuracy")
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import MinMaxScaler
```

```python
model = Pipeline([
    ("Scaling",MinMaxScaler()),
    ("KNN",KNeighborsClassifier(n_neighbors=7))
])
```

```python
model.fit(X_train,y_train)
```

```python
print(classification_report(y_test,model.predict(X_test)))
```

```python
import seaborn as sns
from sklearn.metrics import confusion_matrix

sns.heatmap(confusion_matrix(y_test,model.predict(X_test)),annot=True)
```

```python
import math
sample_size = 500
pop_stdev = house_df["SalePrice"].std()
intervals = []
sample_means = []

for sample in range(25):
    sample = np.random.choice(a=house_df["SalePrice"],size=sample_size)
    sample_mean = sample.mean()
    sample_means.append(sample_mean)

    # get the population standard deviation
    z_critical = stats.norm.ppf(q=0.975)

    # margin of error
    margin_of_error = z_critical * (pop_stdev / math.sqrt(sample_size))
    
    conf_interval = (sample_mean - margin_of_error,
                    sample_mean + margin_of_error)
    print(conf_interval)

    intervals.append(conf_interval)
```

```python
#plt.figure(figsize=(13,9))

yerr_values = [abs(top-bot)/2 for top,bot in intervals]

plt.errorbar(x=np.arange(0.1,25,1),y=sample_means,yerr=yerr_values,fmt='o')

plt.hlines(xmin=0,xmax=25,
          y = house_df["SalePrice"].mean(), linewidth=2.0, color="red")

plt.title('Confidence Intervals for 25 Trials', fontsize=20)
```
